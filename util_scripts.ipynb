```python
file_names = []
for subdir, dirs, files in os.walk(PROCESSED_DIR):
    dirs.sort()
    files.sort()
    for file in files:
        if file == 'file.toml' or file == 'clef2018-full-corpus.txt':
            continue
        file_names.append(file)

print(len(file_names))
with open(META_DIR + 'clef2018-full-corpus.txt', 'w') as f:
    for idx, file in enumerate(file_names):
        if idx == len(file_names) - 1:
            f.write('[none] ' + file)
        else:
            f.write('[none] ' + file + '\n')
```

```python
for subdir, dirs, files in os.walk(DATA_DIR):
    dirs.sort()
    site_name = subdir.split('/')[-1] # site name is the last directory in path
    for file in files:
        page_name = '/'.join([site_name, file])
        file_full_path = os.path.join(subdir, file)
        try:
            site_text = ''
            with open(file_full_path, 'r') as html_file:
                soup = BeautifulSoup(html_file, 'html.parser')
                site_text = soup.get_text()

            with open(PROCESSED_DIR + '/' + file, 'w') as f:
                f.write(site_text)
        except UnicodeDecodeError:
            site_text = ''            
            with open(PROCESSED_DIR + '/' + file, 'w') as f:
                f.write(site_text)
            
            continue
```
